```python
import requests 
import json
import pprint
import pandas as pd
from yelpapi import YelpAPI
from pandas.io.json import json_normalize
from re import search
from config import api_key
```

```python
# define params for search
terms = ['mexican', 'italian']
states = ['CA',
         'VA',
         'OR',
         'VA',
         'NC',
         'FL']

# format api key
yelp_api = YelpAPI(api_key)

# iterate over defined params for api call
for x, i in zip(terms, states):
    # define search terms, sort by ratings, and limit responses to top 20
    data = yelp_api.search_query(term=x , location=i, sort_by='rating', limit=20)
# print the api call in more readable format
print(json.dumps(data, indent=4, sort_keys=True))
```

```python
# check dict keys for response
data.keys()
```

```python
# normalize json response to transform into pandas dataframe
df = json_normalize(data, 'businesses')
df.head()
```

```python
len(df)
```

```python
# select only neccessary columns
df2 = df[['name','categories', 'rating', 'review_count']]
df2.head()
```

```python
# check dtypes
df2.dtypes
```

```python
# clean categories column by selecting only type of restaurant
r_type = []
italian = 'italian'

for italain in df2['categories']:
    if italian == italian:
        r_type.append('italian')
    else:
        r_type.append('mexican')
    
df2['type'] = r_type

df2.head()
```

```python
# find average rating by type
average_rating = df2.groupby('type')['rating'].mean()

average_rating
```

```python
# find total reviews
reviews = df2.groupby('type')['review_count'].sum()
reviews
```
